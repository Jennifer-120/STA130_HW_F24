Question 1

The theoretical Simple Linear Regression model is a statistical model for understanding the relationship between the predictor and outcome variables, which are denoted as X and Y respectively in the mathematical representation. The intercept of the regression line (β0) represents the expected outcome variable when the predictor variable is zero and the slope coefficient (β1) describes how the outcome variable changes with respect to the predictor variable. The core of the theoretical SLR is that it captures the variation with an error term which describes the general population relationship. It is assumed to be normally distributed with mean zero and some standard deviation.

```python
# Question 1

import numpy as np1
import plotly.graph_objects as go
from scipy.stats import uniform, norm

# Parameters for the theoretical model
n = 50          # Number of data points
beta0 = 5       # Theoretical intercept
beta1 = 2       # Theoretical slope
sigma = 2       # Standard deviation for noise

# Generate predictors (Xi) from a uniform distribution
X = uniform.rvs(loc=0, scale=10, size=n)  # X values between 0 and 10

# Generate errors (ei) from a normal distribution
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Create the theoretical Y values using the simple linear model
Y_theoretical = beta0 + beta1 * X  # Without noise
Y_observed = Y_theoretical + errors  # With noise

# Plotting the theoretical line and observed data points using Plotly
fig = go.Figure()

# Add the theoretical linear model line (without noise)
fig.add_trace(go.Scatter(x=X, y=Y_theoretical, mode='lines', name='Theoretical Model (Y = 5 + 2X)', line=dict(color='blue')))

# Add the observed data points (with noise)
fig.add_trace(go.Scatter(x=X, y=Y_observed, mode='markers', name='Observed Data', marker=dict(color='red')))

# Customize the plot
fig.update_layout(
    title="Theoretical Simple Linear Regression Model with Simulated Data",
    xaxis_title="X",
    yaxis_title="Y",
    showlegend=True
)

# Show plot
fig.show(renderer="png")
```

```python
# Question 2

import pandas as pd
import plotly.graph_objects as go
import statsmodels.formula.api as smf

# Assuming X and Y_observed are already defined from the theoretical model

# Convert data to a DataFrame for statsmodels
data = pd.DataFrame({'X': X, 'Y_observed': Y_observed})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y_observed ~ X', data=data).fit()

# Get the fitted values
Y_fitted = model.fittedvalues

# Print the summary of the fitted model
print(model.summary())

# Plotting the observed data points and fitted line
fig = go.Figure()

# Add the observed data points (with noise)
fig.add_trace(go.Scatter(x=X, y=Y_observed, mode='markers', name='Observed Data', marker=dict(color='red')))

# Add the fitted line from the statsmodels regression
fig.add_trace(go.Scatter(x=X, y=Y_fitted, mode='lines', name='Fitted Model', line=dict(color='green', dash='dash')))

# Customize the plot
fig.update_layout(
    title="Fitted Simple Linear Regression Model",
    xaxis_title="X",
    yaxis_title="Y",
    showlegend=True
)

# Show plot
fig.show(renderer="png")
```

```python
# Question 3
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import uniform, norm
import statsmodels.formula.api as smf

# Parameters for the theoretical model
n = 50          # Number of data points
beta0 = 5       # Theoretical intercept
beta1 = 2       # Theoretical slope
sigma = 2       # Standard deviation for noise

# Generate predictors (Xi) from a uniform distribution
X = uniform.rvs(loc=0, scale=10, size=n)  # X values between 0 and 10

# Generate errors (ei) from a normal distribution
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Create the theoretical Y values using the simple linear model
Y_theoretical = beta0 + beta1 * X  # Without noise
Y_observed = Y_theoretical + errors  # With noise

# Convert data to a DataFrame for statsmodels
data = pd.DataFrame({'X': X, 'Y_observed': Y_observed})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y_observed ~ X', data=data).fit()

# Get the fitted values
Y_fitted = model.fittedvalues

# Print the summary of the fitted model
print(model.summary())

# Plotting the theoretical line, observed data points, and fitted line
fig = go.Figure()

# Add the theoretical linear model line (without noise)
fig.add_trace(go.Scatter(x=X, y=Y_theoretical, mode='lines', name='Theoretical Model (Y = 5 + 2X)', line=dict(color='blue')))

# Add the observed data points (with noise)
fig.add_trace(go.Scatter(x=X, y=Y_observed, mode='markers', name='Observed Data', marker=dict(color='red')))

# Add the fitted line from the statsmodels regression
fig.add_trace(go.Scatter(x=X, y=Y_fitted, mode='lines', name='Fitted Model', line=dict(color='green', dash='dash')))

# Customize the plot
fig.update_layout(
    title="Theoretical Model vs. Fitted Model with Simulated Data",
    xaxis_title="X",
    yaxis_title="Y",
    showlegend=True
)

# Show plot
fig.show(renderer="png")
```

The aim of the ordinary least squares (OLS) regression is to find the line that best fits the observed data by minimizing the sum of squared residuals for the purpose of reducing squared errors. A residual is the difference between the Y from sampled data and the predicted Y from the model. Squaring the residuals ensures that they are positive and they won't cancel out in the process of obtaining the sum as well as prevent large errors in the regression line. This would often lead to the regression line to not exactly fit the majority of the datapoints.The red lines on the example model below shows the residuals or how far each point is from the OLS line.

```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Parameters for the theoretical model
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5

# Generate data
x = stats.uniform(x_min, x_range).rvs(size=n)  # Uniformly distributed x values
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)  # Normally distributed errors
Y = beta0 + beta1 * x + errors  # Linear model with noise

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': Y})  # Make sure the column names are consistent

# Fit the model using statsmodels
model_data_specification = smf.ols("y ~ x", data=df)  # Use 'y' as column name here
fitted_model = model_data_specification.fit()  # Fit the model

# Add a column to the DataFrame for legend purposes
df['Data'] = 'Data'

# Plotting the data points and the fitted model
fig = px.scatter(df, x='x', y='y', color='Data', trendline='ols', title='Y vs. x')

# Add the fitted trendline manually
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues, 
                line=dict(color='blue'), name="OLS Trendline")

# Add the theoretical line (orange dotted line)
x_range_vals = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range_vals
fig.add_scatter(x=x_range_vals, y=y_line, mode='lines', 
                name=f'{beta0} + {beta1} * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]], 
                    y=[fitted_model.fittedvalues[i], df['y'][i]], 
                    mode='lines', 
                    line=dict(color='red', dash='dash'), 
                    showlegend=False)

# Add a horizontal line for the mean of y (y-bar)
fig.add_scatter(x=x_range_vals, y=[df['y'].mean()] * 2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

# Show the plot
fig.show(render='png')
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf
from scipy import stats

# Create a mask for long wait times
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Number of bootstrapped samples
n_bootstrap = 1000
bootstrapped_slope_coefficients = []

# Perform bootstrapping
for _ in range(n_bootstrap):
    # Generate a bootstrapped sample
    bootstrap_sample = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    
    # Fit the linear regression model
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    
    # Collect the slope coefficient
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Visualize the bootstrapped slope coefficients
fig = px.histogram(x=bootstrapped_slope_coefficients, 
                   title="Bootstrapped Sampling Distribution of Slope Coefficients", 
                   labels={'x': 'Slope Coefficient'})
fig.show(render="png")

# Parameters for the null hypothesis simulation
beta0_sim = 1.65
beta1_sim = 0  # Null hypothesis assumption of no linear relationship
sigma_sim = 0.37

# Number of simulations
n_simulations = 1000
simulated_slope_coefficients = []

# Perform simulations
for _ in range(n_simulations):
    # Create a simulated dataset with no linear association
    old_faithful_simulation = old_faithful[long_wait].copy()
    old_faithful_simulation['duration'] = beta0_sim + beta1_sim * old_faithful_simulation['waiting'] + \
                                          stats.norm(loc=0, scale=sigma_sim).rvs(size=long_wait.sum())
    
    # Fit the linear regression model to the simulated data
    model_sim = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    
    # Collect the slope coefficient
    simulated_slope_coefficients.append(model_sim.params['waiting'])

# Visualize the simulated sampling distribution of slope coefficients
fig = px.histogram(x=simulated_slope_coefficients, 
                   title="Simulated Sampling Distribution of Slope Coefficients (Null Hypothesis)", 
                   labels={'x': 'Slope Coefficient'})
fig.show(render="png")

# 95% Confidence interval from bootstrapped slope coefficients
bootstrap_conf_interval = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print(f"95% Bootstrapped Confidence Interval for Slope: {bootstrap_conf_interval}")

# Check if 0 is within the 95% CI
contains_zero = 0 >= bootstrap_conf_interval[0] and 0 <= bootstrap_conf_interval[1]
print(f"Does the 95% CI contain 0? {'Yes' if contains_zero else 'No'}")

# Calculate p-value from simulated slope coefficients under the null hypothesis
observed_slope = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params['waiting']
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print(f"Simulated p-value: {simulated_p_value}")

# Compare with the p-value from the original model
original_p_value = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().pvalues['waiting']
print(f"Original p-value from model summary: {original_p_value}")
```

```python
# Question 13

import numpy as np
import pandas as pd
import seaborn as sns

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')
short_wait = old_faithful[old_faithful['waiting'] < 68]['duration']
long_wait = old_faithful[old_faithful['waiting'] >= 68]['duration']

# Observed mean difference
observed_diff = long_wait.mean() - short_wait.mean()

# Number of permutations
n_permutations = 10000
permuted_diffs = []

# Permutation test
combined = np.concatenate([short_wait, long_wait])
for _ in range(n_permutations):
    np.random.shuffle(combined)
    perm_short = combined[:len(short_wait)]
    perm_long = combined[len(short_wait):]
    permuted_diffs.append(perm_long.mean() - perm_short.mean())

# Calculate p-value
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))
print("Permutation Test p-value:", p_value)

# Number of bootstrap samples
n_bootstrap = 10000
bootstrapped_mean_diffs = []

# Bootstrapping within each group
for _ in range(n_bootstrap):
    boot_short = np.random.choice(short_wait, size=len(short_wait), replace=True)
    boot_long = np.random.choice(long_wait, size=len(long_wait), replace=True)
    bootstrapped_mean_diffs.append(boot_long.mean() - boot_short.mean())

# 95% Confidence Interval
ci = np.quantile(bootstrapped_mean_diffs, [0.025, 0.975])
print("95% Bootstrap Confidence Interval for Mean Difference:", ci)
```
